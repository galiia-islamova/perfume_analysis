```python
import pandas as pd

import os
os.chdir(r"C:\Users\user\Desktop\perfume")

```

```python
perfumes = pd.read_csv ("tables/products_perfume.csv")
perfumes.head()
```

```python
suppliers = pd.read_csv ("tables/suppliers_perfume.csv")
suppliers.head()
```

```python
inventory = pd.read_csv ("tables/inventory_perfumes.csv")
inventory.head()
```

```python
sales = pd.read_csv ("tables/sales_perfume.csv")
sales = sales.sort_values('sale_date').reset_index(drop=True)
#change sale id by date
sales['sale_id'] = range(1, len(sales) + 1)
sales.tail(20)
```

```python
# convert date
sales['sale_date'] = pd.to_datetime(sales['sale_date'])
inventory['last_restock_date'] = pd.to_datetime(inventory['last_restock_date'])

# merge
full_clean_data = sales.merge(perfumes, on="perfume_id", how="left")\
.merge(inventory, on="perfume_id", how="left") \
.merge(suppliers, on="supplier_id", how="left")
full_clean_data.head()
```

```python
full_clean_data.to_csv("full_clean_data.csv", index=False)
```
